```python
import coba as cb
import matplotlib.pyplot as plt

L = ['evaluator_id','specs1','specs2','specs3','ssl','neg','ws','er','lr','nr']

res = cb.Result.from_file("out9b.log").filter_fin(l=L,p=['pid','rng'])
res.plot_learners('index','auc',l=L,p=['pid','rng'],span=3)

#self supervised learning only seems to help when reducing dimensions
#this can be seen when we compare the performance of untrained SSL features
#to trained SSL features at width 74 (i.e., evaluator 2 to 3) and the same
#comparison at width 20 (i.e., evaluator 0 to 1).

# self supervised learning plus supervised learning still doesn't perform as well
# as supervised learning only, even at lower dimensions (e.g., evaluator 4 to 0)

# does self-supervised learning ever help ???? Maybe when using a very very small training set?

# one thing we don't have are invariances... Many of the self-supervised learning papers talk
# about the importance of data-augmentation + SSL which I'm not doing. In the case of images
# data-augmentation means things like rotating images or zooming-in/zooming-out. That is the
# image is still the same (aka invariant) even when rotated.
```
